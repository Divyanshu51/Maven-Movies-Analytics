# Merging Tables for Analysis Part
Join films  film_category  category

```python
film_cat = pd.merge(film_category_df, category_df, on="category_id")
film_details = pd.merge(films_df, film_cat, on="film_id")
```

 Join rentals with customers and inventory

```python
rental_data =pd.merge(rental_df,inventory_df, on = "inventory_id")
rental_customer = pd.merge(rental_data,customer_df, on = "customer_id")
rental_full= pd.merge(rental_customer,films_df, on = "film_id")
```

# EDA

**Most Rented Categories**




```python
plt.figure(figsize=(10,6))
film_details['name'].value_counts().head(10).plot(kind='bar', color='mediumseagreen')
plt.title('Top Rented Film Categories')
plt.ylabel('Number of Rentals')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Most Popular Films by Rental Count**

```python
plt.figure(figsize=(10,6))
rental_full['title'].value_counts().head(10).sort_values().plot(kind='barh', color='skyblue')
plt.title('Most Rented Films')
plt.xlabel('Number of Rentals')
plt.ylabel('Title')
plt.tight_layout()
plt.show()
```

**Revenue by Film**

```python
film_revenue = rental_full.groupby('title')['rental_rate'].sum().sort_values(ascending=False).head(10)

# Plot with titles as labels
plt.figure(figsize=(10,6))
film_revenue.plot(kind='barh', color='skyblue')  # horizontal bar makes long titles more readable
plt.xlabel('Total Rental Revenue')
plt.title('Top Revenue-Generating Films')
plt.gca().invert_yaxis()  # Highest on top
plt.tight_layout()
plt.show()
```

**Customer Behaviour**

```python
# Group by full customer ID or full name for uniqueness
top_customers = rental_full.groupby(['customer_id', 'first_name', 'last_name']).size().sort_values(ascending=False).head(10)

# Create formatted labels like "John Smith"
top_customers.index = top_customers.index.map(lambda x: f"{x[1]} {x[2]}")

# Plot
plt.figure(figsize=(10,6))
top_customers.sort_values().plot(kind='barh', color='orange')
plt.title('Top 10 Renting Customers')
plt.xlabel('Number of Rentals')
plt.ylabel('Customer Name')
plt.tight_layout()
plt.show()
```
